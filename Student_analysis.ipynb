```python
import pandas as pd
data = pd.read_csv("student.csv")
print(data.head())
```

```python
print(data.columns)
```

```python
data["Total_Marks"] = data["Math"]+data["Physics"]+data["English"]+data["Computer"]
```

```python
data["Average_Marks"] = data["Total_Marks"]/4
```

```python
data["Attendance (%)"] = (data["Classes_Attended"] / data["Total_Classes"]) * 100
```

```python
def assign_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "P"
    else:
        return "F"
```

```python
data["Grade"] = data["Average_Marks"].apply(assign_grade)
```

```python
topper = data.loc[data["Total_Marks"].idxmax()]
```

```python
pass_count = data[data["Grade"] != "F"].shape[0]
fail_count = data[data["Grade"] == "F"].shape[0]
```

```python
print("Student Performance Analysis")
print(data)
```

```python
data = data.drop(columns=[
    "Total marks",
    "Average marks",
    "Total_marks",
    "Average_marks"
], errors = "ignore")

print("\nFinal Cleaned Data:\n")
print(data)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(data["Name"], data["Average_Marks"])

plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Average Marks Comparison")
plt.xticks(rotation=60)

plt.show()
```

```python
pass_count = data[data["Grade"] != "F"].shape[0]
fail_count = data[data["Grade"] == "F"].shape[0]

plt.figure()
plt.pie([pass_count, fail_count], labels=["Pass", "Fail"], autopct="%1.1f%%")

plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
plt.figure()
plt.bar(data["Name"], data["Attendance (%)"])

plt.xlabel("Students")
plt.ylabel("Attendance Percentage")
plt.title("Attendance Comparison")
plt.xticks(rotation=45)

plt.show()
```
